### Remove null & non-english rows

```python
df.dropna(subset=['user_name', 'date', 'text'], inplace=True)
```

```python
df.size
```

Removed around 51 empty user names and 2000 ish empty dates

```python
# Remove non-english text using WhatTheLang (https://github.com/indix/whatthelang)
wtl = WhatTheLang()
L = []

for row in tqdm(df['text']):
    if len(row) != 0:
        L.append(wtl.predict_lang(row))
    else:
        L.append(None)
        
df['lang'] = L
df.head()
```

```python
df_english = df[df['lang'] == 'en']
```

```python
df_english.head()
```

```python
df_english.drop(['lang'], axis=1, inplace=True)
```

```python
df_english.head()
```

### Limit number of tweets #02

```python
df_grouped_2 = df_english.groupby('date').head(500)
```

```python
# Create checkpoint file
df_grouped_2.to_csv('/kaggle/working/outputs/tweets_small_filtered_cleaned.csv')
```
